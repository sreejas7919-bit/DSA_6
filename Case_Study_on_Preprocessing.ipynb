```python
#Import Libraries
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split


#pandas → Dataset load ചെയ്യാനും DataFrame operations ചെയ്യാനും ഉപയോഗിക്കുന്നു.
#numpy → Numerical calculations ചെയ്യാൻ ഉപയോഗിക്കുന്നു.
#SimpleImputer → Missing values fill ചെയ്യാൻ.
#MinMaxScaler → Numerical features scale ചെയ്യാൻ.
#train_test_split → Dataset training, testing ആയി split ചെയ്യാൻ.
```

```python
#load dataset
df = pd.read_csv("House_Pricing.csv")
```

Duplicate Removal
Rows: Check for duplicate rows in the dataset, if any, and remove them.
Columns: Identify and drop duplicate columns, if any, based on their values.

```python
#Duplicate Rows Count
df.duplicated().sum()
```

```python
#Remove Duplicate Rows
df = df.drop_duplicates()
```

```python
#Duplicate Rows Count
df.duplicated().sum()
```

```python
#Remove Duplicate Columns
df = df.loc[:, ~df.T.duplicated()]
```

```python
print(df.T.duplicated().sum())
```

Handling Missing Values
Identify missing values in each column.
Handle missing values:
For numerical columns, use imputation techniques like mean/median imputation.
For categorical columns, fill with mode.
Document your approach for each feature with missing data.

```python
#Handling Missing Values
#Numerical Columns
num_cols = df.select_dtypes(include=['int64','float64']).columns
df[num_cols]
```

```python
#Categorical Columns
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols]
```

```python
#Mean Imputation
num_imputer = SimpleImputer(strategy='mean')
df[num_cols] = num_imputer.fit_transform(df[num_cols])
df[num_cols]
```

```python
#Mode Imputation
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
df[cat_cols]
```

```python
#Check Missing Values Again
df.isnull().sum()
```

Scaling Numerical Variables
Identify all numerical columns (excluding the target variable SalePrice).
Scale these features using techniques like Min-Max scaling or Standard scaling.

```python
#Scaling,Exclude Target Variable
features_to_scale = num_cols.drop("Sale Price")
df[features_to_scale]
```

```python
#Create Scaler,Min-Max Scaler object create ചെയ്യുന്നു.
scaler = MinMaxScaler()
```

```python
#Apply Scaling
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])
df[features_to_scale]
```

Encoding Categorical Variables
Identify all categorical columns in the dataset.
Apply appropriate encoding techniques:
Use One-Hot Encoding for nominal categories.
Use Label Encoding for ordinal categories, if applicable.

```python
#Encoding
# Re-identify categorical columns to ensure we're working with current columns
cat_cols = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```python
#Outlier Removal
numeric_columns = df.select_dtypes(include=np.number).columns
numeric_columns = numeric_columns.drop("Sale Price")


#SalePrice ഒഴിവാക്കി എല്ലാ numerical columns select ചെയ്യുന്നു.
```

```python
for col in numeric_columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

Train-Test Split
Set aside the SalePrice column as the target variable.
Split the dataset into training (80%) and testing (20%) sets using the train_test_split function from sklearn

```python
#train-Test Split
# Features and Target
X = df.drop("Sale Price", axis=1)

y = df["Sale Price"]
```

```python
#Split Dataset
x_train, x_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

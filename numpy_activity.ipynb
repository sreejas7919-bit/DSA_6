Question-1

Create a NumPy array with values from 10 to 49. Reshape into 4×10 matrix and extract last column.

Question-2

Create two 3×3 matrices, add them, multiply element-wise, find transpose.

Solution-2

FunctionPurposeimport numpy as npImports the NumPy librarynp.full((3,3),5)Creates a 3×3 matrix filled with 5np.arange(1,10)Creates an array from 1 to 9reshape(3,3)Converts array into a 3×3 matrixA + BAdds two matricesA * BPerforms element-wise multiplication.TFinds the transpose of a matrixprint()Displays the output

```python
import numpy as np

# Matrix filled with 5s
A = np.full((3, 3), 5)

# Matrix with values 1 to 9
B = np.arange(1, 10).reshape(3, 3)

# Addition
add_result = A + B

# Element-wise multiplication
mul_result = A * B

# Transpose
transpose_result = mul_result.T

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nAddition:")
print(add_result)

print("\nElement-wise Multiplication:")
print(mul_result)

print("\nTranspose:")
print(transpose_result)
```

Question-3

3. Create a NumPy array of 15 random integers between 10 and 100.
● Replace all values greater than 50 with -1
● Count how many values were replaced.


Solution-3

FunctionPurposeimport numpy as npImports the NumPy librarynp.random.randint(10,101,15)Generates 15 random integers from 10 to 100arr > 50Checks values greater than 50np.sum(arr > 50)Counts values greater than 50arr[arr > 50] = -1Replaces values greater than 50 with -1print()Displays the output

```python
import numpy as np

# Create 15 random integers between 10 and 100
arr = np.random.randint(10, 101, 15)

print("Original Array:")
print(arr)

# Count values greater than 50
count = np.sum(arr > 50)

# Replace values greater than 50 with -1
arr[arr > 50] = -1

print("\nModified Array:")
print(arr)

print("\nNumber of values replaced:", count)
```

Question-4

4. Generate a 5×4 array with random values between 0 and 1 (float).
Find the row with the maximum sum of values.
● Normalize each row using the formula:

Normalized Row=row/sum of row

Solution-4

FunctionPurposeimport numpy as npImports the NumPy librarynp.random.rand(5,4)Creates a 5×4 array of random float valuesnp.sum(arr, axis=1)Calculates the sum of each rownp.argmax(row_sums)Finds the row with the maximum sumrow_sums[:, np.newaxis]Converts row sums into a column vectorarr / row_sums[:, np.newaxis]Normalizes each rowprint()Displays the output

```python
import numpy as np

# Generate 5x4 random float array
arr = np.random.rand(5, 4)

print("Original Array:")
print(arr)

# Sum of each row
row_sums = np.sum(arr, axis=1)

# Row with maximum sum
max_row = np.argmax(row_sums)

print("\nRow Sums:")
print(row_sums)

print("\nRow with Maximum Sum:", max_row)

# Normalize each row
normalized = arr / row_sums[:, np.newaxis]

print("\nNormalized Array:")
print(normalized)
```

Question-5

arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
● Find all unique elements
● Find the duplicates (elements that appear more than once)

Solution-5

FunctionPurposeimport numpy as npImports the NumPy librarynp.array()Creates a NumPy arraynp.unique(arr)Finds unique elementsfor loopIterates through unique elementsnp.count_nonzero(arr == i)Counts occurrences of an elementduplicates.append(i)Adds duplicate elements to the listprint()Displays the output

```python
arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
```

```python
import numpy as np

arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])

# Unique elements
unique_elements = np.unique(arr)

# Find duplicates
duplicates = []

for i in unique_elements:
    if np.count_nonzero(arr == i) > 1:
        duplicates.append(i)

print("Array:")
print(arr)

print("\nUnique Elements:")
print(unique_elements)

print("\nDuplicate Elements:")
print(duplicates)
```

Question-6

create a 6×6 NumPy array with values from 1 to 36.
● Extract the middle 4×4 submatrix.
● Then extract all even numbers from this submatrix.

Solution-6

Function
-------Purpose
-------import numpy as npImports the NumPy librarynp.arange(1,37)Creates an array from 1 to 36reshape(6,6)Converts the array into a 6×6 matrixarr[1:5,1:5]Extracts the middle 4×4 submatrixsubmatrix % 2 == 0Checks for even numberssubmatrix[submatrix % 2 == 0]Extracts even numbersprint()Displays the output

```python
import numpy as np

# Create 6x6 array
arr = np.arange(1, 37).reshape(6, 6)

# Extract middle 4x4 submatrix
submatrix = arr[1:5, 1:5]

# Extract even numbers
even_numbers = submatrix[submatrix % 2 == 0]

print("Original Array:")
print(arr)

print("\nMiddle 4x4 Submatrix:")
print(submatrix)

print("\nEven Numbers:")
print(even_numbers)
```

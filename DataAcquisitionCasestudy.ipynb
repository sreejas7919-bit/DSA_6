Task 8
Merge all four datasets into one DataFrame.
Hint:
● Use the common column (country)
● Choose an appropriate join method.

```python
merged_data = population.merge(gdp, on='country', how='outer')

merged_data.head()
```

```python
merged_data = merged_data.merge(internet_users, on='country', how='outer')

merged_data.head()
```

```python
literacy_rate.rename(columns={'name':'country'}, inplace=True)
merged_data = merged_data.merge(literacy_rate, on='country', how='outer')

merged_data.head()
```

```python
merged_data.shape
```

```python
merged_data.columns
```

Task 9
Create the following calculated columns:
1. Internet Penetration (%)
2. GDP Per Capita
Hint:
Round the calculated values to two decimal places.

```python
merged_data['Internet Penetration (%)'] = (
    merged_data['internet_users'] / merged_data['population']
) * 100

merged_data['Internet Penetration (%)'] = merged_data['Internet Penetration (%)'].round(2)
merged_data[['country','Internet Penetration (%)']].head()
```

```python
merged_data['GDP Per Capita'] = (
    merged_data['GDP'] / merged_data['population']
)

merged_data['GDP Per Capita'] = merged_data['GDP Per Capita'].round(2)
```

```python
merged_data[['country','GDP Per Capita']].head()
```

```python
merged_data.head()
```

Task 10

```python
import sqlite3
```

```python
connection = sqlite3.connect("country_statistics.db")
```

```python
pd.read_sql("""
SELECT country, gdp
FROM country_statistics
ORDER BY gdp DESC
LIMIT 5;
""", connection)
```

```python
pd.read_sql("""
SELECT AVG(literacy_rate) AS average_literacy_rate
FROM country_statistics;
""", connection)
```

```python
pd.read_sql("""
SELECT country, literacy_rate
FROM country_statistics
WHERE literacy_rate > 90;
""", connection)
```

```python
pd.read_sql("""
SELECT country, `Internet Penetration (%)`
FROM country_statistics
ORDER BY `Internet Penetration (%)` DESC
LIMIT 5;
""", connection)
```

```python
pd.read_sql("""
SELECT SUM(population) AS total_population
FROM country_statistics;
""", connection)
```

```python
pd.read_sql("""
SELECT country, `GDP Per Capita`
FROM country_statistics
ORDER BY `GDP Per Capita` DESC
LIMIT 1;
""", connection)
```

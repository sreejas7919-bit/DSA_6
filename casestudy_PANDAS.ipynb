**CASE STUDY PANDAS**

1. Import pandas and numpy

```python
import pandas as pd
import numpy as np
```

2) Load dataset using pd.read_csv()  


```python
df = pd.read_csv('/content/IMDB-Movie-Data.csv')
df
```

3. Display shape, head, column names, datatypes, missing values

```python
print("Shape:")
print(df.shape)

print("\nFirst 5 Rows:")
print(df.head())

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

4. Check and remove duplicates

```python
print("Duplicate Rows:", df.duplicated().sum())
df = df.drop_duplicates()
print("After Removing Duplicates:", df.shape)
```

5. Rename column (Title → Movie_Title)

```python
df.rename(columns={"Title":"Movie_Title"}, inplace=True)

print(df.columns)
```

6)  Get unique genres


```python
genres = set()

for genre in df["Genre"]:
    for g in genre.split(","):
        genres.add(g.strip())

print(genres)
```

7. Extract Number of Genres per Movie







```python
df["Genre_Count"] = df["Genre"].apply(lambda x: len(x.split(",")))

print(df[["Movie_Title", "Genre_Count"]].head())
```

8)  Handle missing values in 'Revenue (Millions)' and 'Metascore





```python
df["Revenue (Millions)"] = df["Revenue (Millions)"].fillna(
    df["Revenue (Millions)"].mean()
)

df["Metascore"] = df["Metascore"].fillna(
    df["Metascore"].median()
)

print("Revenue Missing Values:",
      df["Revenue (Millions)"].isnull().sum())

print("Metascore Missing Values:",
      df["Metascore"].isnull().sum())
```

```python
df["Revenue (Millions)"].fillna(
    df["Revenue (Millions)"].mean(),
    inplace=True
)

df["Metascore"].fillna(
    df["Metascore"].mean(),
    inplace=True
)

print(df.isnull().sum())
```

9. Top 5 Movies by Rating

```python
top5_movies = df.nlargest(5, "Rating")
print(top5_movies[["Movie_Title", "Rating"]])
```

10) Movies with highest/lowest revenue





```python
highest = df.loc[df["Revenue (Millions)"].idxmax()]

lowest = df.loc[df["Revenue (Millions)"].idxmin()]

print("Highest Revenue Movie:")
print(highest["Movie_Title"])

print("\nLowest Revenue Movie:")
print(lowest["Movie_Title"])
```

11)  Avg. rating by Director


```python
avg_rating = df.groupby("Director")["Rating"].mean()

print(avg_rating)
```

12) Movies per year


```python
df["Year"].value_counts()
```

13) Movies with rating > 8 & revenue > 100M


```python
result = df[
    (df["Rating"] > 8) &
    (df["Revenue (Millions)"] > 100)
]

print(result[["Movie_Title", "Rating", "Revenue (Millions)"]])
```

14) Group by Director: avg Rating, Revenue


```python
df.groupby("Director").agg({
    "Rating":"mean",
    "Revenue (Millions)":"mean"
})
```

15) Group by Year: count, avg Rating


```python
df.groupby("Year").agg({
    "Movie_Title":"count",
    "Rating":"mean"
})
```

16) Create a new column Profitability:

   a) Revenue divided by Duration
   b) Find top 10 most profitable movies

```python
df["Profitability"] = df["Revenue (Millions)"] / df["Runtime (Minutes)"]

print(df[["Movie_Title", "Profitability"]].head())
```

17) Create a column that categorizes movies by rating:
a) Rating < 6: Poor
b) 6 ≤ Rating < 7.5: Average
c) Rating ≥ 7.5: Excellent

```python
def category(rating):
    if rating < 6:
        return "Poor"
    elif rating < 7.5:
        return "Average"
    else:
        return "Excellent"

df["Rating_Category"] = df["Rating"].apply(category)

print(df[["Movie_Title", "Rating", "Rating_Category"]].head())
```

18) Create a custom function to score movies based on:
i)
Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
Apply a function to dataset to assign the score

```python
def movie_score(row):
    return row["Rating"] * 10 + row["Revenue (Millions)"]

df["Movie_Score"] = df.apply(movie_score, axis=1)

print(df[["Movie_Title", "Movie_Score"]].head())
```

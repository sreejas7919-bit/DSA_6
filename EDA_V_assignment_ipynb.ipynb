# **Exploratory Data Analysis - Visualizations**

# Dataset: Tips
**Restaurant bills, tips, gender analysis**

```python
# Import libraries


import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
tips = sns.load_dataset("tips")
```

```python
#head
print(tips.head())
```

```python
# Shape
print("Shape:", tips.shape)
```

```python
# information
print(tips.info())
```

```python
# Statistical Summary of this data set

print(tips.describe())
```

```python
# Missing values

print(tips.isnull().sum())
```

```python
#count plot

plt.figure(figsize=(6,4))
sns.countplot(x='sex', data=tips)
plt.title("Count of Male and Female Customers")
plt.show()



#insight : Male customers are more than female customers.
```

```python
#HISTOGRAM :  a special type of bar graph that shows how often numbers appear within certain ranges

plt.figure(figsize=(6,4))
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.show()


#Most customers have a total bill between $10 and $20.
```

```python
#Box Plot :a visual graph that shows the spread and distribution of a dataset by dividing it into four equal parts

plt.figure(figsize=(6,4))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Total Bill by Day")
plt.show()

#INSIGHT : Saturday has the highest bill amounts and some outliers.
```

```python
#Scatter Plot : a graph that uses dots to show the relationship between two different variables.

plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.title("Total Bill vs Tip")
plt.show()


# Insight: Customers with higher bills generally give higher tips.
```

```python
# Heatmap : a data visualization that uses color to show numerical values across a grid, map, or webpage

plt.figure(figsize=(6,4))
corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Insight: Total bill and tip have a strong positive correlation.
```

```python
# BAR PLOT : visual graph that uses rectangular bars to compare different categories of data

plt.figure(figsize=(6,4))
sns.barplot(x='day', y='tip', data=tips)
plt.title("Average Tip by Day")
plt.show()

#insights :Customers give higher average tips on Saturday and Sunday.
```

```python
#Count Plot (Smoker)
#specific type of bar graph that counts how many times each category appears in a dataset.

plt.figure(figsize=(6,4))
sns.countplot(x='smoker', data=tips)
plt.title("Smoker vs Non-Smoker")
plt.show()

#insights : Non-smokers are more than smokers
```

```python
#Step Plotconnects data points using horizontal and vertical steps instead of straight lines.

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,5,3,7,6]

plt.step(x, y)
plt.title("Step Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Stem Plot displays data as vertical lines (stems) with markers at the top.

x = [1,2,3,4,5]
y = [2,5,3,7,6]

plt.stem(x, y)
plt.title("Stem Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Subplot displays multiple graphs in a single figure.

x = [1,2,3,4]
y1 = [2,4,6,8]
y2 = [1,3,5,7]

plt.subplot(1,2,1)
plt.plot(x, y1)
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x, y2)
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
#pair plot : a grid of multiple mini-graphs that lets you see relationships across an entire dataset at once

sns.pairplot(tips, hue='sex')
plt.show()

#insight : 1.Total bill and tip show a positive relationship,
          #2. Male and female customers have similar data distributions.
```

```python
sns.pairplot(tips)
plt.show()

# insight : Higher total bill → Higher tip.
#customers with higher total bills usually give higher tips, and larger groups tend to have higher bills.
```

QUESTION 1

```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/titanic_dataset (1).csv')
df.head(10)
```

QUESTION 2
Find the number of missing values in each column.

| Function        | Purpose                                            |
| --------------- | -------------------------------------------------- |
| `pd.read_csv()` | Loads the CSV file into a DataFrame                |
| `df.isnull()`   | Identifies missing (null) values                   |
| `.sum()`        | Counts the number of missing values in each column |
| `print()`       | Displays the output                                |


```python
missing_values = df.isnull().sum()
print(missing_values)
```

QUESTION 3
Find the percentage of passengers who survived, grouped by gender.

```python
df.groupby('Sex')['Survived'].mean() * 100
```

QUESTION 4
Find the average, median, and standard deviation of passenger ages.

```python
average_age = df['Age'].mean()
median_age = df['Age'].median()
std_dev_age = df['Age'].std()
print("Average Age:", average_age)
print("Median Age:", median_age)
print("Standard Deviation of Age:", std_dev_age)
```

question 5
Count how many passengers were in each class (pclass).

```python
class_counts = df['Pclass'].value_counts()
print(class_counts)
```

question 6
Show all passengers who were female, under age 18, and survived.

```python
result = df[(df['Sex'] == 'female') &
            (df['Age'] < 18) &
            (df['Survived'] == 1)]

print(result)
```

q 7
Replace missing age values with the median age.

```python
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
print(df['Age'])
```

Q8 . Create a column age_group with values:
a. 'Child' for age < 18
b. 'Adult' for age 18–60
c. 'Senior' for age > 60
d. Use: apply() or np.where() or custom function  


```python
def age_group(age):
    if age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['Age'].apply(age_group)

df[['Age','age_group']].head()
```

Q9
Display the top 5 passengers who paid the highest fare.

```python
df.sort_values(by='Fare', ascending=False).head(5)
```

Q10
 Find the average fare paid, grouped by both pclass and sex.


```python
avg_fare = df.groupby(['Pclass', 'Sex'])['Fare'].mean()
print(avg_fare)
```
